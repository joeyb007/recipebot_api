The final step before training all data is to vectorize inputs, so that XGBoost can actually make sense of them. This will be done using an NLP technique known as TF-IDF, which converts words to numbers based on their frequency in a givewn doument relative to their frequency of ocurrance in all other documents

```python
# Begin by creating dataframe

import pandas as pd
import ast

df = pd.read_csv("../CSV_data/seed_labeled_XGBoost_training_data.csv")

# Edit ingredients to be one string (rather than a list)
def concat_ingredients(ingredientlist):
    return " ".join(ast.literal_eval(ingredientlist))

concatenated_ingredients = df["ingredients"].apply(concat_ingredients)

df["ingredients_text"] = concatenated_ingredients

display(df)
```
