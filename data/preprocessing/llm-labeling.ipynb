Now that our XGBoost models have been created, we can use these models to robustly label more examples and fine-tune an LLM for recipe generation. This will esentially act as a three-step pipeline.
1. Recipe data will be preprocessed into a suitable output format and labeled with XGB 
2. The LLM will be locally fine-tuned with this data
3. LLM outputs will be judged by XGBoost models, effectively acting as an offline RL-style learning system.
The present document addresses the first part of this problem.
